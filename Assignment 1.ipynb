```python
import pandas as pd

df = pd.read_csv('ks-projects-201801.csv')

display(df.head())
```

```python
for col in df.columns:
    print(col)
    print(df[col].unique())
```

```python
# Remove 'other' states in 'state' (cancelled)
# Transform target category to binary value (Sucess or failed)
df = df[df['state'].isin(['successful', 'failed'])]

# Separate the target variable "state" from the features
target = df['state']
features = df.drop('state', axis=1)

# Remove information not deemed relevant 
# Category - Is a subset of main category
# ID - Not relevent to research question
features = features.drop(['category', 'ID'], axis=1)
```

```python
display(features.head(30))
display(target.head())
```

```python
# The pledged, usd pledged and usd_pledged_real columns all seem to overlap.
# The description of the features are described as: 

#usd_pledged: conversion in US dollars of the pledged column (conversion done by kickstarter).
#usd pledge real: conversion in US dollars of the pledged column (conversion from Fixer.io API).
#usd goal real: conversion in US dollars of the goal column (conversion from Fixer.io API).

# This is slightly confusing, however I believe that usd_pledged_real describes the final amount pledged in usd.
# Therefore we will drop the other columns to simplify the data with best respect to the research question.
features = features.drop(['pledged', 'usd pledged'], axis=1)
```

```python
display(features.head())
```

```python
for col in features.columns:
    print(col)
    print(features[col].unique())
```

```python
# A potential feature that could be useful would be to calculate how long did the crowdfund go for? NOTE: This could also be used instead due to TIME exclusion requirement in specs
# convert the "launched" and "deadline" columns to datetime format
features['launched'] = pd.to_datetime(features['launched'])
features['deadline'] = pd.to_datetime(features['deadline'])

# calculate the duration of the project in days and store in a new column called "duration"
features['duration'] = (features['deadline'] - features['launched']).dt.days
```
